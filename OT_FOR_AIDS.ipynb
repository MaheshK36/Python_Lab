```python
import pulp
lp = pulp.LpProblem("LPP", pulp.LpMaximize)
x = pulp.LpVariable('x', lowBound=0)
y = pulp.LpVariable('y', lowBound=0)
lp += 5*x+4*y
lp += 6*x+4*y<=24
lp += x+2*y<=6
lp.solve()
print("x=",x.value(),",y=",y.value(),",Maximize Z=",pulp.value(lp.objective))
```

```python
import numpy as np
from scipy.optimize import linprog
cost = [2,4,5,3]
A_eq = [[1,1,0,0],[0,0,1,1]]
b_eq = [20,30]
result = linprog(c=cost, A_eq=A_eq, b_eq=b_eq)
print("Minimum Transportation Cost =",result.fun)
```

```python
from scipy.optimize import linear_sum_assignment
cost = [
    [9, 2, 7, 8],
    [6, 4, 3, 7],
    [5, 8, 1, 8],
    [7, 6, 9, 4]
]
row_ind, col_ind = linear_sum_assignment(cost)
min_cost = sum(cost[i][j] for i, j in zip(row_ind, col_ind))
print("Optimal Assignments:\n", list(zip(row_ind, col_ind)))
print("Minimum Assignment Cost:", min_cost)
```

```python
from scipy.optimize import linprog
c=[-3,-2]
A=[ [1, 1],[1, 0],[0, 1] ]
b = [4, 2, 3]
x_bounds = [0, None]
y_bounds = [0, None]
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds])
print("LP Relaxation Solution:", result.x)
print("Objective Value:", -result.fun)
```

```python
def knapsack(weights, values, capacity):
    n = len(values)
    dp = [0] * (capacity + 1)
    for i in range(n):
        for w in range(capacity, weights[i] - 1, -1):
            dp[w] = max(dp[w],values[i] + dp[w - weights[i]])
    return dp[capacity]
weights = [12, 15, 30, 20, 25]
values = [5, 4, 10, 6, 8]
capacity = 60
print(knapsack(weights, values, capacity))
```
